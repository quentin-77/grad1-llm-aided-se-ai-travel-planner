```python
cur_time=$(date +%s)
params_base64=$(echo -n '{"engine_type":"sms16k","aue":"raw"}' | base64 | tr -d '\n')
checksum=$(printf "%s" "YOUR_API_KEY${cur_time}${params_base64}" | md5)
curl https://api.xfyun.cn/v1/service/v1/iat \
  -H "Content-Type: application/x-www-form-urlencoded; charset=utf-8" \
  -H "X-Appid: YOUR_APPID" \
  -H "X-CurTime: ${cur_time}" \
  -H "X-Param: ${params_base64}" \
  -H "X-CheckSum: ${checksum}" \
  --data-urlencode "audio=$(base64 < your-test.wav)"
```
